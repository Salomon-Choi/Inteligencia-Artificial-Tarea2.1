```python
#separara las caracteristicas
#caracteristicas de entrada

X = datos_na.drop(['median_house_value'], axis=1)

#etiquera de salida o repuesta
Y= datos_na['median_house_value']
```

```python
#separa el conjunto de caracteristicas en datos de prueba y entrenamiento

from sklearn.model_selection import train_test_split
X_t, X_te,  Y_t, Y_te = train_test_split(X,Y,test_size=0.2)
```

```python
Y_t.shape #solo pa ver que onda
```

```python
#entrenar el modelo
from sklearn.linear_model import LinearRegression

modelo=LinearRegression()
modelo.fit(X_t, Y_t)
```

```python
#hacer predicciones :D
p=modelo.predict(X_te)
p
```

```python
#

pdf=pd.DataFrame(p)
comparar={
    'Predicciones':p,
    'valor original':Y_te
}

pd.DataFrame(comparar)
```

```python
modelo.score(X_t, Y_t)
```

```python
modelo.score(X_te, Y_te)
```

```python
#calcular para ver cuanto me desvie
import numpy as np
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(Y_te,p)
#valor promedio del cual me aleje de la predic
mse=np.sqrt(mse)
mse
```
